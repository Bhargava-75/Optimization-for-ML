**Question 11.** Outlier-robust Huber regression

In a linear regression problem, we are given data $({\bf x}_i,y_i)\in \mathbb{R}^n \times \mathbb{R}$, $i=1,\ldots, m$. We fit an affine model

$$\hat{y}_i = {\bf x}_i^T{\boldsymbol \beta}$$

for $i=1,\ldots, m$ with ${\boldsymbol \beta} \in \mathbb{R}^n$.


In standard linear regression, we choose ${\boldsymbol \beta}$ by solving 
$$
\underset{{\boldsymbol \beta}}{\text{minimize}} 
\quad \sum\limits_{i=1}^m r_i^2 =  \sum\limits_{i=1}^m (y_i - {\bf x}_i^T{\boldsymbol \beta})^2. 
$$

Huber regression is a technique that is robust to outliers. In Huber regression, we replace the square function with the *Huber function*

$$
\phi(u) = \left\{ \begin{array}{ll}  u^2 & |u| \leq M\\
2M|u|  -  M^2 & |u| > M
\end{array}\right.
$$
where $M>0$ is the Huber threshold. That is, we solve

$$
\underset{{\boldsymbol \beta}}{\text{minimize}} \sum\limits_{i=1}^m \phi(y_i - {\bf x}_i^T{\boldsymbol \beta})
$$
for variable ${\boldsymbol \beta}.$

**[5 pts] Question 11b.** Plot the Huber function $\phi(u)$ for different values of $M \in \{1,5, 1000\}$ with $u\in [-10,10]$. Compare with the plot of the quadratic function $u^2$.

```python
# Write your code here

import numpy as np
from matplotlib import pyplot as plt

# Define the Huber loss
def Phi(u,M):
  if(abs(u)<=M):
    return u**2
  else:
    return 2*abs(u)*M - M**2

plt.subplot(2, 1, 1)
u = np.linspace(-10,10,100)
M1=[1]*100
M2=[5]*100
M3=[1000]*100
phi_with_M_1 =list(map(Phi,u,M1))
phi_with_M_2 =list(map(Phi,u,M2))
phi_with_M_3 =list(map(Phi,u,M3))

plt.plot(u,phi_with_M_1, color='R', label='M=1')
plt.plot(u,phi_with_M_2, color='G', label='M=5')
plt.plot(u,phi_with_M_3, color='B', label='M=1000')
plt.plot(u,u**2, color='black', label='U^2')

plt.xlabel("u")
plt.ylabel("(u)")
plt.title("Huber function")
plt.legend()

plt.show()
plt.subplot(2, 1, 2)
plt.xlabel("u")
plt.ylabel("u^2")
plt.title("Quadratic function")
plt.plot(u,u**2, color='black', label='U^2')
plt.legend()
plt.show()

# By compare the plot U^2 with Huber Function when M = 1000 and U^2 is same because |U| is never greater than the M so it is same as Linear Regression
```

We generate data with $m=500$ measurements and $n=300$ regressors.
We randomly choose ${\boldsymbol \beta}$ and ${\bf x}_i \sim \mathcal N(0,I)$.
We set $y_i = {\bf x}_i^T{\boldsymbol \beta} + \epsilon_i$, where $\epsilon_i \sim
\mathcal N(0,1)$. 
To introduce outliers, with probability $p$ we replace $y_i$ with $-y_i$. We generate $50$ problem instances, with $p$ varying from $0$ to $0.15$,


* Recover ${\boldsymbol \beta} \in {\bf R}^n$ from the measurements 
$y\in {\bf R}^m$ and compare the three approaches: linear regression, Huber regression and *clairvoyant* regression. In clairvoyant regression we know which measurements had their sign flipped.

* Plot the relative error 
$$
\frac{\|\hat{\boldsymbol \beta} - {\boldsymbol \beta}\|_2}{\|{\boldsymbol \beta}\|2},
$$
where $\hat{\boldsymbol \beta}$ is the recovered ${\boldsymbol \beta}$.

* Comment on your observations


Use [sklearn.linear_model](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.linear_model) for Huber regression and linear regression


```python
## Solution 11c.
import numpy as np
from sklearn.linear_model import HuberRegressor, LinearRegression
from sklearn.preprocessing import StandardScaler
## Generate data 
np.random.seed(1)
n = 300
m = 450
TESTS = 50

beta_true = 5*np.random.normal(size=(n,1))
X = np.random.normal(0,1,(n,m))
Y = np.zeros((m,1))
p_vals = np.linspace(0,0.15, num=TESTS)

# Linear regression
lr = LinearRegression()

# Huber regression
hr = HuberRegressor()

#Errors
lr_errors = []
hr_errors = []
cr_errors = []

epsilon = np.random.normal(0,1, size=(m,1))

#Function to compute relative error
def relative_error(beta_cap, beta_true):
    return np.linalg.norm(beta_cap - beta_true) / np.linalg.norm(beta_true)

for tst, p in enumerate(p_vals):
    # Generate the sign changes.
    factor = 2*np.random.binomial(1, 1-p, size=(m,1)) - 1
    index=[]
    for i in range(m):
      if(factor[i]==1):
        index.append(i)
    X_new=[]
    for i in index:
      X_new.append(X[:,i])
    X_modify = np.array(X_new)
    Y_modify= X_modify.dot(beta_true)
    Y = factor*X.T.dot(beta_true) + epsilon
    # Solve standard regression 
    lr.fit(X.T,Y)
    lr_errors.append(relative_error(lr.coef_, beta_true))

    # Solve clairvoyant regression
    lr.fit(X_modify,Y_modify)
    cr_errors.append(relative_error(lr.coef_, beta_true))
    # Solve Huber regression  with M=1
    hr.fit(X.T,Y)
    hr_errors.append(relative_error(hr.coef_,beta_true))
```

```python
import matplotlib.pyplot as plt
# Plot the relative error for the three methods
# for p_vals
p_values = np.linspace(0, 0.15, 50)
plt.plot(p_values, lr_errors, label='Linear regression')
plt.plot(p_values, hr_errors, label='Huber regression')
plt.plot(p_values, cr_errors, label='Clairvoyant regression')
plt.xlabel('Proportion of outliers')
plt.ylabel('Relative error')
plt.legend()
plt.show()
```

Our aim is to build a classifier to detect the *Iris virginica* type based on the petal width and length features.


Given data $({\bf x}_i,y_i)$, $i=1,\ldots, m$. The ${\bf x}_i \in {\bf R}^n$ are feature vectors and $y_i \in \{0, 1\}$ indicates the associated boolean classes.

Our goal is to construct a linear classifier determined by the halfspace ${\bf x}^T{\boldsymbol \beta}>0$. Specifically, we predict that the $i$th instance belongs to class 1 if ${\bf x}^T{\boldsymbol \beta} >0$ and class $0$ otherwise.
We find ${\boldsymbol \beta}$ by solving the maximizing likelihood problem

$$
\underset{\boldsymbol \beta}{\text{maximize}} \quad f({\boldsymbol \beta}) = \sum_{i=1}^{m} y_i {\bf x}_i^T{\boldsymbol \beta} - \log(1 + \exp ({\bf x}_i^T{\boldsymbol \beta})).
$$

Sure! Here is the same proof written in LaTeX format:

We want to show that the function $f(\beta) = \sum_{i=1}^m y_i x_i^T \beta - \log(1 + \exp(x_i^T \beta))$ is $L$-smooth, i.e., its gradient is Lipschitz continuous with constant $L$.

The gradient of $f(\beta)$ is given by:

$$\nabla f(\beta) = \sum_{i=1}^m y_i x_i - \sum_{i=1}^m \frac{1}{1+\exp(x_i^T \beta)} \exp(x_i^T \beta) x_i$$

We can simplify the second term by multiplying and dividing by $\exp(x_i^T \beta)$:

$$\nabla f(\beta) = \sum_{i=1}^m y_i x_i - \sum_{i=1}^m \frac{\exp(x_i^T \beta)}{1+\exp(x_i^T \beta)} \cdot \frac{1}{\exp(x_i^T \beta)} \exp(x_i^T \beta) x_i$$

Using the fact that $\frac{1}{1+\exp(x)} \leq 1$ for all $x$, we can bound the second term as:

$$\left| \sum_{i=1}^m \frac{\exp(x_i^T \beta)}{1+\exp(x_i^T \beta)} \cdot \frac{1}{\exp(x_i^T \beta)} \exp(x_i^T \beta) x_i \right| \leq \sum_{i=1}^m |x_i|$$

Therefore, we can bound the norm of the gradient as:

$$\left|\nabla f(\beta)\right| \leq \sum_{i=1}^m |x_i| + \left|\sum_{i=1}^m y_i x_i\right|$$

Next, let's compute the Hessian of $f(\beta)$:

$$H(f)(\beta) = \sum_{i=1}^m \frac{\exp(x_i^T \beta)}{(1+\exp(x_i^T \beta))^2} x_i x_i^T$$

Using the fact that $0 \leq \frac{\exp(x)}{1+\exp(x)} \leq 1$ for all $x$, we can bound the eigenvalues of the Hessian as:

$$0 \leq \lambda_i(H(f)(\beta)) \leq |x_i|^2$$

Therefore, the Lipschitz constant of the gradient of $f(\beta)$ is bounded by:

$$L = \max_i |x_i|^2$$

Thus, we have shown that $f(\beta)$ is $L$-smooth with Lipschitz constant $L = \max_i |x_i|^2$.

[sklearn.linear_model](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html#sklearn.linear_model.LogisticRegression) to train a logistic regression model.


Show the linear boundary displaying two features petal width and length along with the instances (i.e., *Iris virginica* in one halfspace and not *Iris virginica* in the other). 








```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np
X = iris["data"][:, (2, 3)]  # petal length, petal width
y = (iris["target"] == 2).astype(np.int) # 1 if Iris virginica, else 0


# Enter your code here
logr = LogisticRegression(solver='lbfgs')
logr.fit(X,y)
b = logr.intercept_[0]
w1, w2 = logr.coef_.T
print(logr.coef_)
# Calculate the intercept and gradient of the decision boundary.
c = -b/w2
m = -w1/w2

# Plot the data and the classification with the decision boundary.
xmin, xmax = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
ymin, ymax = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
xd = np.array([xmin, xmax])
yd = m*xd + c
plt.plot(xd, yd, 'k', lw=1)
plt.fill_between(xd, yd, ymin, color='skyblue')
plt.fill_between(xd, yd, ymax, color='lightgreen')

plt.scatter(*X[y==0].T, s=8, linewidths = 2,label ="Iris-Setosa & Iris-Versicolour")
plt.scatter(*X[y==1].T, s=8,linewidths = 2, label = "Iris virginica")
plt.xlim(xmin, xmax)
plt.ylim(ymin, ymax)
plt.ylabel(r'Petal Length')
plt.xlabel(r'Petal Width')
plt.legend()
plt.show()


```
