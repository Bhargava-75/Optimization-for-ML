```python
import numpy as np
import matplotlib.pyplot as plt

#Creating two contour plots
feature_x = np.linspace(-5.0, 5.0, 70)
feature_y = np.linspace(-7.0, 7.0, 70)
[X, Y] = np.meshgrid(feature_x, feature_y)
fig, ax = plt.subplots(1, 3,figsize=(25, 5))
Z = 0.1 * X ** 2 + 2 * Y ** 2
ax[0].contourf(X, Y, Z)
ax[1].contourf(X, Y, Z)
ax[2].contourf(X, Y, Z)

#Derivative function
def grad_fun(x1,x2):
  return [0.2*x1 , 4*x2]


# Gradient Descent
def Gradient_Descent(eta,intial_point):
  normal_grad_points=[]
  normal_grad_points.append(intial_point)
  for i in range(20):
    x1=intial_point[0]
    x2=intial_point[1]
    derivate=[eta*j for j in grad_fun(x1,x2)]
    intial_point=[intial_point[j]-derivate[j] for j in range(2)]
    normal_grad_points.append(intial_point)
  return normal_grad_points

#Heavy Ball Method
def Heavy_Ball_Method(intial_point,eta,beta):
  heavy_ball_points=[]
  heavy_ball_points.append(intial_point)
  for i in range(20):
    x1=intial_point[0]
    x2=intial_point[1]
    derivate=[eta*j for j in grad_fun(x1,x2)]
    if(i==0):
      next_point=[intial_point[j]-derivate[j] for j in range(2)]
      previous_point=intial_point
      intial_point=next_point
    else:
      next_point=[intial_point[j]-derivate[j]+beta*(intial_point[j]-previous_point[j]) for j in range(2)]
      previous_point=intial_point
      intial_point=next_point
    heavy_ball_points.append(intial_point)
  return heavy_ball_points



intial_point=[4,6]
eta=0.4
normal_grad_points=Gradient_Descent(eta,intial_point)
x_points=[i[0] for i in normal_grad_points]
y_points=[i[1] for i in normal_grad_points]

# plots filled contour plot with gradient descent.

ax[0].set_title('Gradient descent with eta=0.4')
ax[0].set_xlabel('X1')
ax[0].set_ylabel('X2')
ax[0].scatter(x_points,y_points,s=10,c='w')
ax[0].plot(x_points,y_points,c='w')
ax[0].scatter([0],[0],c='skyblue')


intial_point=[4,6]
eta=0.5
normal_grad_points=Gradient_Descent(eta,intial_point)
x_points=[i[0] for i in normal_grad_points]
y_points=[i[1] for i in normal_grad_points]

# plots filled contour plot with gradient descent.

ax[1].set_title('Gradient descent with eta=0.5')
ax[1].set_xlabel('X1')
ax[1].set_ylabel('X2')
ax[1].scatter(x_points,y_points,s=10,c='w')
ax[1].plot(x_points,y_points,c='w')
ax[1].scatter([0],[0],c='skyblue')

intial_point=[4,6]
eta=0.5
beta=0.3
heavy_ball_points=Heavy_Ball_Method(intial_point,eta,beta)
x_points=[i[0] for i in heavy_ball_points]
y_points=[i[1] for i in heavy_ball_points]

# plots filled contour plot with heavy-ball method.
ax[2].set_title('Heavy Ball Method with eta=0.5,beta=0.3')
ax[2].set_xlabel('X1')
ax[2].set_ylabel('X2')
ax[2].scatter(x_points,y_points,s=10,c='w')
ax[2].plot(x_points,y_points,c='w')
ax[2].scatter([0],[0],c='skyblue')

plt.show()

#By Observing the graphs for each case Heavy Ball method converges very fastly to the minimum point. And we have to pick right eta value , if not it will never converges
# are it takes so much time to converge like in graph2
```

**[20 pts] Question 7.** Outlier-robust Huber regression.

In a Huber regression problem, we are given data $({\bf x}_i,y_i)\in \mathbb{R}^n \times \mathbb{R}$, $i=1,\ldots, m$. we solve

$$
\underset{{\boldsymbol \beta}}{\text{minimize}} \sum\limits_{i=1}^m \phi(y_i - {\bf x}_i^T{\boldsymbol \beta})
$$
for variable ${\boldsymbol \beta},$ where the *Huber function*

$$
\phi(u) = \left\{ \begin{array}{ll}  u^2 & |u| \leq M\\
2M|u|  -  M^2 & |u| > M
\end{array}\right.
$$
where $M>0$ is the Huber threshold. 


**Implementation**

We generate data with $m=500$ measurements and $n=300$ regressors.
We randomly choose ${\boldsymbol \beta}$ and ${\bf x}_i \sim \mathcal N(0,I)$.
We set $y_i = {\bf x}_i^T{\boldsymbol \beta} + \epsilon_i$, where $\epsilon_i \sim
\mathcal N(0,1)$. 
To introduce outliers, with probability $p$ we replace $y_i$ with $-y_i$. We generate a problem instance with $p= 0.01$.


* Implement gradient descent, to recover ${\boldsymbol \beta} \in {\bf R}^n$ from the measurements $y\in {\bf R}^m$ by solving Huber regression.

* Plot the suboptimality gap
$$
f({\boldsymbol \beta}_k) - f^\star,
$$
where you can use the solution from homework 1, i.e., [sklearn.linear_model](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.linear_model) to compute $f^\star$. 

* Use a fixed step size of $1/L$. 

* Suppose now you have an incorrect estimate $L$ and use that to compute the step size. Based on the numerical results, comment on your observations how critical is to use the correct value for $L$.






```python
## Solution 6.
import numpy as np
from sklearn.linear_model import HuberRegressor
import matplotlib.pyplot as plt
## Generate data 
np.random.seed(1)
n = 300
m = 450

beta_true = 5*np.random.normal(size=(n,1))
X = np.random.randn(m, n)
Y = np.zeros((m,1))

#Huber function
def Huber_fun(U,M):
  result = np.where(np.abs(U)>M,2*M*np.abs(U)-M**2,U**2)
  return np.sum(result)/m


#Gradient of Huber function 
def Grad_Huber_fun(U,M):
  return -np.dot(X.T, U)/m
```

```python
#Setting parameters for problem
p=0.01
epsilon=np.random.randn(m,1)
Y_true = np.dot(X,beta_true)+epsilon
factor = 2*np.random.binomial(1,1-p,size=(m,1))-1
Y_out = Y_true*factor

iterations = 1000
axis = np.linspace(0,iterations-1,iterations)
M=1


## Huber regression from sklearn
huber_model = HuberRegressor(epsilon=M,max_iter=7000)
huber_model.fit(X,Y_out)
beta_huber = huber_model.coef_
f_opt = Huber_fun(Y_out- np.dot(X,beta_huber).reshape(-1,1)+ huber_model.intercept_,M)
print(f_opt)
```

```python
#Here L is 2 so eta = 1/2 = 0/5
#Huber regression Using gradient descent method with correct eta value 0.5
eta=0.5
beta_values = np.zeros((iterations,n))
fun_vals_huber_gd = np.zeros((1,iterations))
fun_vals_huber_gd[0,0] = Huber_fun(Y_out-np.dot(X,beta_values[0,:]).reshape(-1,1),M)
for i in range(1,iterations):
  beta_values[i,:] = beta_values[i-1,:] - eta*Grad_Huber_fun(Y_out-np.dot(X,beta_values[i-1,:]).reshape(-1,1),M)[:,0]
  fun_vals_huber_gd[0,i] = Huber_fun(Y_out-np.dot(X,beta_values[i-1,:]).reshape(-1,1),M)
plt.semilogy(axis,np.abs(fun_vals_huber_gd[0,:]-f_opt))
plt.xlabel('Iterations')
plt.ylabel('Suboptimality gap')
plt.xlim([0,998])
plt.legend(['$\eta = 0.5$'])
plt.figure()
```

```python
#Huber regression Using gradient descent  but with wrong eta value=0/75
eta=0.75
beta_values = np.zeros((iterations,n))
fun_vals_huber_gd = np.zeros((1,iterations))
fun_vals_huber_gd[0,0] = Huber_fun(Y_out-np.dot(X,beta_values[0,:]).reshape(-1,1),M)
for i in range(1,iterations):
  beta_values[i,:] = beta_values[i-1,:] - eta*Grad_Huber_fun(Y_out-np.dot(X,beta_values[i-1,:]).reshape(-1,1),M)[:,0]
  fun_vals_huber_gd[0,i] = Huber_fun(Y_out-np.dot(X,beta_values[i-1,:]).reshape(-1,1),M)
plt.semilogy(axis,np.abs(fun_vals_huber_gd[0,:]-f_opt))
plt.xlabel('Iterations')
plt.ylabel('Suboptimality gap')
plt.xlim([0,998])
plt.legend(['$\eta = 0.75$'])
plt.figure()
```

**[20 pts] Question 8.** 

Consider the problem of minimizing a piecewise linear function 

$$
{\rm minimize}_{\bf x} \,\, f({\bf x}) = \underset{i=1,\ldots,m}{\max} ({\bf a}_i^T{\bf x} + b_i)
$$

with variable ${\bf x} \in \mathbb{R}^d$. We implement the subgradient method with $d=20$ variables and $m=100$ terms. We generate data ${\bf a}_i$ and $b_i$ from a unit normal distribution and initialize with ${\bf x}_0 = {\bf 0}$.






Implement a subgradient method to solve the above problem.  Express the above problem as a linear problem and use 
[CVX solver](https://www.cvxpy.org/examples/index.html) to find $f^{\rm opt}$.


Through simulations, plot the convergence of $f^{\rm best}_k - f^{\rm opt}$ for

1. fixed step size $\eta = 0. 01$
2. Polyak's step size rule
3. Heavy-ball method: ${\bf x}_{k+1} = {\bf x}_{k} - \eta_k {\bf g}_{k} + \beta_k({\bf x}_{k} - {\bf x}_{k-1})$, where $\eta_k = 1/k$ (also try Polyak's step size) and $\beta \in [0,1]$. Try a few values for $\beta$ and report your observations. 









```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt

## Generate data 
np.random.seed(1)
m=100
d=20
A = np.random.randn(m, d)
b = np.random.randn(m, 1)



#Finding optimal value using cvxpy library
c1 = np.ones(d)
A1 = np.hstack((A,-np.ones((m,1))))
c = np.append(np.zeros(d),1).reshape(-1,1)
x_cvx = cp.Variable((d+1,1))
prob = cp.Problem(cp.Minimize(c.T@x_cvx), [A1@x_cvx <= -b])
prob.solve()
f_opt = prob.value
print('Optimal value using cvxpy is:',f_opt)

#Gradient Computation
def Grad_Computation(A, b, x):
  d = len(x)
  gt = np.matmul(A, x) + b
  idx = np.argmax(gt)
  grad = A[idx,:].T
  grad = grad.reshape(d,1)
  return grad, idx

#
def SG_Constant_Step(A, b, x_0, eta, nIter):
  eta1 = eta
  x_sg = x_0 
  f_i = []
  f_sg = []
  for i in range(nIter):
    [grad, idx] = Grad_Computation(A, b, x_sg) 
    eta = eta1
    x_sg = x_sg - eta*grad # update x_sg
    f = min(max(np.matmul(A, x_sg) + b)) 
    f_i.append(f)
    f_sg.append(min(f_i)) 
  return f_sg, x_sg


def SG_Polyak(A, b, x_0, nIter):
  x_plk = x_0 
  f_plk_i = []
  f_plk = []
  for i in range(nIter):
    [grad, idx] = Grad_Computation(A, b, x_plk) 
    f = min(max(np.matmul(A, x_plk) + b))
    eta = (f- f_opt)/np.linalg.norm(grad)**2
    x_plk = x_plk - eta*grad 
    f = min(max(np.matmul(A, x_plk) + b)) 
    f_plk_i.append(f)
    f_plk.append(min(f_plk_i)) 
  return f_plk, x_plk


def SG_diminishing(A, b, x_0, beta):
  d = len(x_0)
  x_hb = x_0
  f = min(max(np.matmul(A, x_hb) + b))
  x_hb_prev = np.zeros((d,1))
  f_hb = [f]
  f_hb_i = [f]
  k = 1
  for i in range(nIter):
    eta = 1/k
    [grad, idx] = Grad_Computation(A, b, x_hb)
    x_new = x_hb - eta*grad + beta*(x_hb - x_hb_prev)
    f = min(max(np.matmul(A, x_new) + b)) 
    f_hb_i.append(f)
    f_hb.append(min(f_hb_i)) 
    x_hb_prev = x_hb
    x_hb = x_new
    k = k + 1
  return f_hb, x_hb

def SG_Dim_Pol(A, b ,x_0, beta):
  d = len(x_0)
  x_hb = x_0
  f = min(max(np.matmul(A, x_hb) + b))
  x_hb_prev = np.zeros((d,1))
  f_hb = [f]
  f_hb_i = [f]
  k = 1
  for i in range(nIter):
    [grad, idx] = Grad_Computation(A, b, x_hb)
    f = min(max(np.matmul(A, x_hb) + b))
    eta = (f- f_opt)/np.linalg.norm(grad)**2
    x_new = x_hb - eta*grad + beta*(x_hb - x_hb_prev)
    f = min(max(np.matmul(A, x_new) + b)) 
    f_hb_i.append(f)
    f_hb.append(min(f_hb_i)) 
    x_hb_prev = x_hb
    x_hb = x_new
    k = k + 1
  return f_hb, x_hb

x_0 = np.zeros((d, 1))
eta = 0.05
nIter = 3000
[f_sg, x_sg] = SG_Constant_Step(A, b, x_0, eta, nIter)
[f_plk, x_plk] = SG_Polyak(A, b, x_0, nIter)
beta = np.array([0.25, 0.5, 0.75])
plt.figure()
for k in range(len(beta)):
  b1 = beta[k]
  [f_hb, x_hb] = SG_diminishing(A, b, x_0, b1)
  plt.semilogy(range(nIter+1), [f_hb[i]-f_opt for i in range(nIter+1)])
plt.legend([r' $\beta = 0.25$', r'$\beta = 0.5$', r'$\beta = 0.75$'])
plt.title('SG with diminishing step size')
plt.xlabel('Iterations')
plt.ylabel('Suboptimality gap')

beta = np.array([0.1, 0.25, 0.5])
plt.figure()
for k in range(len(beta)):
  b1 = beta[k]
  [f_hb_plk, x_hb_plk] = SG_Dim_Pol(A, b ,x_0, b1)
  plt.semilogy(range(nIter+1), [f_hb_plk[i]-f_opt for i in range(nIter+1)])
plt.legend([r' $\beta = 0.1$', r'$\beta = 0.25$', r'$\beta = 0.5$'])
plt.title('SG with Polyak step size and momentum')
plt.xlabel('Iterations')
plt.ylabel('Suboptimality gap')
plt.figure()
plt.semilogy(range(nIter), [f_sg[i]-f_opt for i in range(nIter)])
plt.semilogy(range(nIter), [f_plk[i]-f_opt for i in range(nIter)])
plt.legend(['SG with constant $\eta$', 'SG with Polyak step size'])
plt.title('SG with constant step size and Polyak step size')
plt.xlabel('Iterations')
plt.ylabel('Suboptimality gap')

```
